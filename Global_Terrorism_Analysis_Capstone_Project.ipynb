## <b> The Global Terrorism Database (GTD) is an open-source database including information on terrorist attacks around the world from 1970 through 2017. The GTD includes systematic data on domestic as well as international terrorist incidents that have occurred during this time period and now includes more than 180,000 attacks. The database is maintained by researchers at the National Consortium for the Study of Terrorism and Responses to Terrorism (START), headquartered at the University of Maryland.</b>

```python
# Replacing the null values for object type

df1["provstate"].fillna("Not present",inplace = True)
df1["city"].fillna("Not present",inplace = True)
df1["location"].fillna("Not present",inplace = True)
df1["summary"].fillna("Not present",inplace = True)
df1["targsubtype1_txt"].fillna("Not present",inplace = True)
df1["corp1"].fillna("Not present",inplace = True)
df1["target1"].fillna("Not present",inplace = True)
df1["natlty1_txt"].fillna("Not present",inplace = True)
df1["motive"].fillna("Not present",inplace = True)
df1["weapsubtype1_txt"].fillna("Not present",inplace = True)
df1["weapdetail"].fillna("Not present",inplace = True)
df1["propextent_txt"].fillna("Not present",inplace = True)
df1["propcomment"].fillna("Not present",inplace = True)
df1["scite1"].fillna("Not present",inplace = True)
df1["scite2"].fillna("Not present",inplace = True)
df1["scite3"].fillna("Not present",inplace = True)
```

# **Exploratory Analysis and Visualization**

```python
# Number of attacks in different years
number_attacks=df1['iyear'].value_counts().sort_index()
number_attacks
```

```python
# Plotting total number of attacks through each year
number_attacks.plot(kind='bar',color='purple')
plt.rcParams['figure.figsize']=(20,7)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.title('Total number of attacks',fontdict=font1)
plt.xlabel('Years',fontdict=font2)
plt.ylabel('Number of attacks',fontdict=font2)
plt.grid()
```

### **From this graph, we have seen that 1971 is the year in which minimum (471) attacks got happened where the maximum attacks (16903) was done in 2014.**

```python
# Finding out which type of weapon was most frequently used
weapon_type=df1['weaptype1_txt'].value_counts().sort_index()
weapon=df1['weaptype1_txt'].unique()
weaptype=(np.sort(weapon))
weapon_type
```
